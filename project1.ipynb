```python
import pandas as pd
import numpy as np
import os
import logging
import time
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Ensure log directory exists
os.makedirs('log', exist_ok=True)

# Setup logging
logging.basicConfig(
    filename="log/ingestion2.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="a"
)

# Create database engine
engine = create_engine('sqlite:///pizza.db')

def ingest_db(df, table_name):
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)

def load_raw_data():
    data_path = os.path.join(os.getcwd(), 'Downloads', 'pizza_sales')
    start = time.time()

    for file in os.listdir(data_path):
        if not file.endswith('.csv'):
            continue
        try:
            file_path = os.path.join(data_path, file)
            df = pd.read_csv(file_path)
            ingest_db(df, file[:-4])
            logging.info(f'Successfully ingested {file} into database')
        except Exception as e:
            logging.error(f'Failed to ingest {file}: {e}')

    end = time.time()
    total_time = (end - start) / 60
    logging.info('----------------------- Ingestion Complete -----------------------')
    logging.info(f'Total Time Taken: {total_time:.2f} minutes')

if __name__ == '__main__':
    load_raw_data()
```
